```python
import numpy as np
import pandas as pd
```

```python
def build_csv(raw_path, final_path, r,l,c,vs, verbose=1):
    df = pd.read_csv(raw_path, sep=r"\s+", engine='python')
    df.columns = ['time','V','IC','IL','IR']
    df["R"] = r
    df['L'] = l
    df['C'] = c
    df['vs'] = vs

    df.to_csv(final_path, index=False)
    if verbose==1:
        print(f'file made, saved to : {final_path}')
```

```python
raw_path = 'raw_texts\\train_6.txt'
final_path = 'train_data\\train_6.csv'
r=10
l=1
c=0.05
vs=1.5
```

```python
build_csv(raw_path,final_path,r,l,c,vs)
```
